### underscore 

`_` - специальная переменная в python. Она используется для сохранения результата последней выполненой команды и, в том числе, для обозначения переменных, которые не будут использоваться:

````python
for _ in range(5):
    print("I can use underscore")
````

При именовании переменных underscore обычно используют для обозначения функций и переменных, которые будут использоватся только внутри класса.

````python
class A():
    def __init__(self):
        self._var = 'su'
        self.__var = 'du'

a = A()
print('a._var = ', a._var)
````

Двойной underscore (`__`) во время создания класса транслируется в `_ИмяКласса__ИмяПеременной`:

````python
try:
    a.__var # AttributeError: 'A' object has no attribute '__var'
except AttributeError:
    print('no such attribute as __var')

print(a._A__var)
````

##### Копирование списков.

Если переменную\_1 приравнять к уже существующему списку (переменная\_2), то в переменную\_1 вместо копии списка попадет ссылка на уже существующий список. В таком случае обе переменные будут указывать на один и тот же объект, а в переменной\_1 хранится ссылка (псевдокопися) на список.

Для создания полной копии нужно использовать метод `copy()` или слайс всего массива.

````python
a = [1, 2, 3]
b = a # псевдокопия
print('a:', a)

print('Изменим один из элементов b, изменится ли a?')
b[1] = 100
print('a:', a)
print('b и a указывают на один объект? Ответ:', b is a)
````

````python
a = [1, 2, 3]

b = a.copy() # копия
print('b =', b, ' - полная копия a')
print('b - это a? Ответ:', b is a, '\n')

b = a[:]
print('b =', b, ' - полная копия a')
print('b - это a? Ответ:', b is a, '\n')

print('Изменим один из элементов b, изменится ли a?')
b[1] = 7
print('b =', b)
print('a =', a)
````

Ни метод `copy()`, ни слайсы, не будут работать с вложенными массивами, для их копирования можно воспользоватся функций `deepcopy()` из модуля `copy`

###### Задача отфильтровать все элементы со значением больше 2.

Способ 1. Стандартный цикл:

````python
a = [1, 2, 3, 4]
b = []
for i in a:
    if i > 2:
        b.append(i)
print(b)
````

Способ 2. Списковое включение:

````python
a = [1, 2, 3, 4]
b = [i for i in a if i > 2]
print(b)
````

Способ 3. Использование функции `filter`:

````python
a = [1, 2, 3, 4]
b = list(filter(lambda x: x > 2, a))
print(b)
````

###### Добавить к каждому элементу 3.

Способ 1. Стандартный цикл:

````python
a = [1, 2, 3]
for i in range(len(a)):
    a[i] += 3
print(a)
````

Способ 2. Списковое включение:

````python
a = [1, 2, 3]
a = [i + 3 for i in a]
print(a)
````

Способ 3. Использование функции `map`:

````python
a = [1, 2, 3]
a = list(map(lambda x: x + 3, a))
print(a)
````

###### get(key [, defval])
Если ключ присутствует в словаре, то метод возвращает значение, соответствующее этому ключу. Если ключ отсутствует, то возвращается значение None или значение из второго параметра

````python
d = {'a': 1, 'b': 2}
print(d.get('a'))
print(d.get('c'))
````

###### setdefault(key [,defval])
Если ключ присутствует в словаре, то метод возвращает значение, соответствующее этому ключу. Если ключ отсутствует, то вставляется новый элемент со значением, указанным во втором параметре. Если второй параметр не указан, то None:

````python
d = {'a': 1, 'b': 2}
print(d.setdefault('a'))

d.setdefault('d', 3)
print(d)
````

###### pop
удаляет элемент с указанным ключом и возвращает его значение. Если ключ отсутствует, то возвращается значение из второго параметра. Если ключ отсутствует и второй параметр не указан, то возбуждается исключение KeyError.

````python
d = {'a': 1, 'b': 2}
d.pop('a')
````

````python
d = {'a': 1, 'b': 2}
d.pop('e', 0) # Попытка достать значение, которого нет в словаре, но с указанием запасного параметра
````

## Декоратор


обертка для других функций или класса.

````python
def makebold(fn):
    def wrapped():
        return '<b>' + fn() + '</b>'
    return wrapped

def makeitalic(fn): 
    def wrapped(): 
        return "<i>" + fn() + "</i>" 
    return wrapped 

@makebold
@makeitalic 
def hello():
    return 'hello world'

hello()
````

## Рекурсия

````python
def factor(n):
    if n <= 1:
        return 1
    else:
        return n * factor(n-1)

factor(3)
````
